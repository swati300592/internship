# 1. Scrape  the  
details  of  
most  viewed  videos  on  
YouTube  
from  Wikipedia.  
Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:  A) 
Rank   
B) Name   
C) Artist   
D) Upload date   
E) Views  

```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

table = soup.find('table', {'class': 'wikitable'})

rank_list = []
name_list = []
artist_list = []
upload_date_list = []
views_list = []

for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    
    if len(columns) >= 5:
        rank = columns[0].text.strip()
        name = columns[1].text.strip()
        artist = columns[2].text.strip()
        upload_date = columns[4].text.strip()
        views = columns[3].text.strip()

        rank_list.append(rank)
        name_list.append(name)
        artist_list.append(artist)
        upload_date_list.append(upload_date)
        views_list.append(views)

for i in range(len(rank_list)):
    print(f"A) Rank: {rank_list[i]}\nB) Name: {name_list[i]}\nC) Artist: {artist_list[i]}\nD) Upload date: {upload_date_list[i]}\nE) Views: {views_list[i]}\n")
```

# 2. Scrape the details team India’s international fixtures from bcci.tv.   
Url = https://www.bcci.tv/.   
You need to find following details:   
A) Series   
B) Place   
C) Date   
D) Time   
Note: - From bcci.tv home page you have reach to the international fixture page through code.  

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bcci.tv/international/fixtures"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

fixtures = soup.find_all("div", class_="js-list")

series_list = []
place_list = []
date_list = []
time_list = []

for fixture in fixtures:
    series = fixture.find("p", class_="fixture__additional-info").text.strip()
    place = fixture.find("span", class_="fixture__location").text.strip()
    date = fixture.find("span", class_="fixture__date").text.strip()
    time = fixture.find("span", class_="fixture__time").text.strip()

    series_list.append(series)
    place_list.append(place)
    date_list.append(date)
    time_list.append(time)

for i in range(len(series_list)):
    print(f"A) Series: {series_list[i]}\nB) Place: {place_list[i]}\nC) Date: {date_list[i]}\nD) Time: {time_list[i]}\n")
```

# 3. Scrape the details of State-wise GDP of India from statisticstime.com.   
Url = http://statisticstimes.com/   
You have to find following details: A) Rank   
B) State   
C) GSDP(18-19)- at current prices   
D) GSDP(19-20)- at current prices   
E) Share(18-19)   
F) GDP($ billion)   
Note: - From statisticstimes home page you have to reach to economy page through code.  

```python
import requests
from bs4 import BeautifulSoup

url = "http://statisticstimes.com/"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

economy_link = soup.find("a", {"href": "/economy/indian-states-gdp.php"})

if economy_link:
    
    economy_url = "http://statisticstimes.com" + economy_link["href"]

    response = requests.get(economy_url)

    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find("table", {"id": "table_id"})

    rank_list = []
    state_list = []
    gsdp_1819_list = []
    gsdp_1920_list = []
    share_1819_list = []
    gdp_billion_list = []
    
    for row in table.find_all("tr")[1:]:
        columns = row.find_all("td")
        rank = columns[0].text.strip()
        state = columns[1].text.strip()
        gsdp_1819 = columns[2].text.strip()
        gsdp_1920 = columns[3].text.strip()
        share_1819 = columns[4].text.strip()
        gdp_billion = columns[5].text.strip()

        rank_list.append(rank)
        state_list.append(state)
        gsdp_1819_list.append(gsdp_1819)
        gsdp_1920_list.append(gsdp_1920)
        share_1819_list.append(share_1819)
        gdp_billion_list.append(gdp_billion)

    for i in range(len(rank_list)):
        print(f"A) Rank: {rank_list[i]}\nB) State: {state_list[i]}\nC) GSDP(18-19) at current prices: {gsdp_1819_list[i]}\nD) GSDP(19-20) at current prices: {gsdp_1920_list[i]}\nE) Share(18-19): {share_1819_list[i]}\nF) GDP($ billion): {gdp_billion_list[i]}\n")
else:
    print("Link to 'GDP of Indian states' not found.")
```

# 4. Scrape the details of trending repositories on Github.com.   
Url = https://github.com/   
You have to find the following details:   
A) Repository title   
B) Repository description   
C) Contributors count   
D) Language used  

```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP GET request to the GitHub Trending page
url = "https://github.com/trending"
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Find the container that holds the list of trending repositories
repositories = soup.find_all("li", class_="col-12 d-block width-full py-4 border-bottom")

# Iterate through the repositories and extract the details
for repo in repositories:
    repo_title = repo.find("h1").text.strip()
    repo_description = repo.find("p", class_="col-9 color-text-secondary my-1 pr-4").text.strip()
    
    # Contributors count and language used are found within different elements
    details = repo.find_all("span", class_="d-inline-block ml-0 mr-3")
    contributors_count = details[0].text.strip()
    language_used = details[1].text.strip()

    print(f"A) Repository title: {repo_title}")
    print(f"B) Repository description: {repo_description}")
    print(f"C) Contributors count: {contributors_count}")
    print(f"D) Language used: {language_used}")
    print("\n")
```

# 5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/  You have to find the 
following details:   
A) Song name   
B) Artist name   
C) Last week rank   
D) Peak rank   
E) Weeks on board   
Note: - From the home page you have to click on the charts option then hot 100-page link through code.  

```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP GET request to the Billboard website
url = 'https://www.billboard.com/'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the "Charts" link and click on it
    charts_link = soup.find('a', text='Charts')
    if charts_link:
        charts_url = charts_link['href']
        charts_response = requests.get(charts_url)

        if charts_response.status_code == 200:
            charts_soup = BeautifulSoup(charts_response.text, 'html.parser')

            # Find the "Hot 100" link and click on it
            hot100_link = charts_soup.find('a', text='Hot 100')
            if hot100_link:
                hot100_url = hot100_link['href']
                hot100_response = requests.get(hot100_url)

                if hot100_response.status_code == 200:
                    hot100_soup = BeautifulSoup(hot100_response.text, 'html.parser')

                    # Extract the details of the top 100 songs
                    song_details = hot100_soup.find_all('li', class_='chart-list-item')

                    for song in song_details:
                        song_name = song.find('span', class_='chart-list-item-title').text
                        artist_name = song.find('div', class_='chart-list-item-artist').text
                        last_week_rank = song.find('div', class_='chart-list-item-last-week').text
                        peak_rank = song.find('div', class_='chart-list-item-peak-rank').text
                        weeks_on_board = song.find('div', class_='chart-list-item-weeks-on-chart').text

                        print("Song:", song_name)
                        print("Artist:", artist_name)
                        print("Last Week Rank:", last_week_rank)
                        print("Peak Rank:", peak_rank)
                        print("Weeks on Board:", weeks_on_board)
                        print()

                else:
                    print("Failed to access the Hot 100 page")
            else:
                print("Hot 100 link not found on the Charts page")
        else:
            print("Failed to access the Charts page")
    else:
        print("Charts link not found on the home page")
else:
    print("Failed to access the Billboard website")
```

# 6. Scrape the details of Highest selling novels.   
A) Book name   
B) Author name   
C) Volumes sold   
D) Publisher   
E) Genre   
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare 

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table')

    if table:

        book_names = []
        author_names = []
        volumes_sold = []
        publishers = []
        genres = []


        rows = table.find_all('tr')
        for row in rows[1:]: 
            columns = row.find_all('td')
            if len(columns) >= 5:
                book_names.append(columns[0].text.strip())
                author_names.append(columns[1].text.strip())
                volumes_sold.append(columns[2].text.strip())
                publishers.append(columns[3].text.strip())
                genres.append(columns[4].text.strip())

        for i in range(len(book_names)):
            print("Book Name:", book_names[i])
            print("Author Name:", author_names[i])
            print("Volumes Sold:", volumes_sold[i])
            print("Publisher:", publishers[i])
            print("Genre:", genres[i])
            print()

    else:
        print("Table not found on the page")

else:
    print("Failed to access The Guardian's URL")

```

# 7. Scrape the details most watched tv series of all time from imdb.com.   
Url = https://www.imdb.com/list/ls095964455/ You have 
to find the following details:   
A) Name   
B) Year span   
C) Genre   
D) Run time   
E) Ratings   
F) Votes  

```python
import requests
from bs4 import BeautifulSoup


url = 'https://www.imdb.com/list/ls095964455/'
response = requests.get(url)


if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    tv_series_list = soup.find_all('div', class_='lister-item-content')

    for series in tv_series_list:

        name = series.find('h3', class_='lister-item-header').find('a').text.strip()
        year_span = series.find('span', class_='lister-item-year').text.strip()
        genre = series.find('span', class_='genre').text.strip()
        run_time = series.find('span', class_='runtime').text.strip()

        rating_element = series.find('div', class_='inline-block ratings-imdb-rating')
        rating = rating_element.strong.text.strip() if rating_element else "N/A"

        votes_element = series.find('span', {'name': 'rk'})
        votes = votes_element.get('data-value') if votes_element else "N/A"

        print("Name:", name)
        print("Year Span:", year_span)
        print("Genre:", genre)
        print("Run Time:", run_time)
        print("Ratings:", rating)
        print("Votes:", votes)
        print()

else:
    print("Failed to access the IMDb URL")
```

# 8. Details of Datasets from UCI machine learning repositories.   
Url = https://archive.ics.uci.edu/  You 
have to find the following details:   
A) Dataset name   
B) Data type   
C) Task   
D) Attribute type   
E) No of instances   
F) No of attribute G) Year   
Note: - from the home page you have to go to the Show All Dataset page through code.  

```python
import requests
from bs4 import BeautifulSoup

url = "https://archive.ics.uci.edu/"

response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.content, "html.parser")
    

    show_all_datasets_link = soup.find("a", text="Show All Datasets")
    
    if show_all_datasets_link:

        show_all_datasets_url = url + show_all_datasets_link["href"]
        

        datasets_response = requests.get(show_all_datasets_url)
        

        if datasets_response.status_code == 200:

            datasets_soup = BeautifulSoup(datasets_response.content, "html.parser")
            
            table = datasets_soup.find("table")
            

            dataset_details = []

    
            for row in table.find_all("tr"):
                columns = row.find_all("td")
                if len(columns) >= 7:
                    dataset_name = columns[0].get_text()
                    data_type = columns[1].get_text()
                    task = columns[2].get_text()
                    attribute_type = columns[3].get_text()
                    no_of_instances = columns[4].get_text()
                    no_of_attributes = columns[5].get_text()
                    year = columns[6].get_text()
                    dataset_details.append((dataset_name, data_type, task, attribute_type, no_of_instances, no_of_attributes, year))

            for details in dataset_details:
                print("Dataset Name:", details[0])
                print("Data Type:", details[1])
                print("Task:", details[2])
                print("Attribute Type:", details[3])
                print("No of Instances:", details[4])
                print("No of Attributes:", details[5])
                print("Year:", details[6])
                print("\n")

        else:
            print("Failed to retrieve the 'Show All Datasets' page.")
    else:
        print("Link to 'Show All Datasets' not found on the home page.")

else:
    print("Failed to access the UCI Machine Learning Repository.")

```
